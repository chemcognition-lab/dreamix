# Multiclass with CE Loss

```python
dataloaders: dict = {"multiclass": get_multiclass_dataset(n_features=25, n_classes=5)}
glm_dict = {
    "multiclass": {
        "input_dim": 25,
        "output_dim": 5,
        "tasktype": "multiclass",
        "loss_fn_index": 0,
    },
}
train_dataloader = dataloaders["multiclass"][0]
test_dataloader = dataloaders["multiclass"][1]
scaler = dataloaders["multiclass"][2]
# plot data (train)
plot_classification_distribution(np.array(train_dataloader.dataset.x), np.array(train_dataloader.dataset.y), title="Multiclass", xlabel="Principal Component 1", ylabel="Principal Component 2")
# plot data (test)
plot_classification_distribution(np.array(test_dataloader.dataset.x), np.array(test_dataloader.dataset.y), title="Multiclass", xlabel="Principal Component 1", ylabel="Principal Component 2")
# run ML model
results = train_loop(glm_dict, dataloaders, epochs=10)
print(results["multiclass"][0])
```
